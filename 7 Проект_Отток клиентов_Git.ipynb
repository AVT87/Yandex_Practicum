В нашей задаче наблюдается сильный дисбаланс классов (англ. class imbalance), что плохо сказывается на обучении модели. 
Классы несбалансированны, когда их соотношение далеко от 1:1. Баланс классов (англ. class balance) наблюдается, если их количество примерно равно.
**Accuracy** не подходит. Нужна новая метрика. Класс с меткой «1» называется положительным, с меткой «0» — отрицательным.
Если сравнить эти ответы с предсказаниями, получается такое деление:
истинно положительные (англ. True Positive, TP) и истинно отрицательные ответы (англ. True Negative, TN);
ложноположительные (англ. False Positive, FP) и ложноотрицательные ответы (англ. False Negative, FN).
Характеристики «положительный» и «отрицательный» относятся к предсказанию, а «истинный» и «ложный» — к его правильности.

Чтобы увеличить качество модели при дисбалансе классов применяются техники:
- взвешивание классов,
- upsampling,
- downsampling.

### Разбивка данных на выборки

Целевой признак у нас категориальный (уход клиента), то решается задача классификации. Когда категорий всего две (ушел клиент, или нет) — речь идёт о бинарной (двоичной) классификации. Значит, данные нужно разбить на три части: обучающую, валидационную и тестовую. Размеры тестового и валидационного наборов обычно равны. Исходные данные разбивают в соотношении 3:1:1.

```python
# создание переменной для признаков
features = df.drop(['exited'], axis=1) # извлекаем признаки

# создание переменной для целевого признака
target = df['exited']

# разделяем данные на обучающую и валидационную выборки в соотношении 60/40 (обучающая/валидационная)
features_train, features_valid, target_train, target_valid, = train_test_split(
    features, target, test_size=0.40, random_state=112)

# разделяем поровну (по 50% от 40% оставшихся данных, или по 20% от исходных) валидационную и тестовую выборку
features_valid, features_test, target_valid,target_test = train_test_split(
    features_valid, target_valid, test_size=0.50, random_state=112)
```

Многие алгоритмы машинного обучения зависят от масштаба признаков. Масштабирование признаков помогает улучшить производительность алгоритма и повысить качество его работы.

Масштабирование признаков в машинном обучении означает приведение значений признаков к одному диапазону. Это может быть выполнено путем нормализации (приведения к диапазону от 0 до 1) или стандартизации (приведения к нормальному распределению со средним значением 0 и стандартным отклонением 1).

### DecisionTreeClassifier

```python
# Создадим модель дерева с перебором параметров

# Диапазон значений depth_values_DTC для перебора в max_depth
depth_values_DTC = range(1, 21)

# Создание списка для сохранения метрики f1
f1_list_valid_DTC = []

# Перебор значений в depth_values_DTC
for depth in depth_values_DTC:
    
    # Создание модели
    model_DTC = DecisionTreeClassifier(random_state=112, max_depth=depth)
    
    # Обучение модели на train выборке
    model_DTC.fit(features_train, target_train)
    
    # Предсказания модели на valid выборке
    pred_valid_DTC = model_DTC.predict(features_valid)
    prob_valid_DTC = model_DTC.predict_proba(features_valid)
    prob_one_valid_DTC = prob_valid_DTC[:, 1]
    
    # Оценка f1 на valid выборке
    f1_valid_DTC = f1_score(target_valid, pred_valid_DTC)
    f1_list_valid_DTC.append(f1_valid_DTC)
    
# Нахождение модели с наилучшим значением f1 на валидационной выборке
best_index_DTC = f1_list_valid_DTC.index(max(f1_list_valid_DTC))
best_f1_DTC = f1_list_valid_DTC[best_index_DTC]
best_depth_DTC = depth_values_DTC[best_index_DTC]
best_model_DTC = DecisionTreeClassifier(random_state=112, max_depth=best_depth_DTC)
best_model_DTC.fit(features_train, target_train)

# предсказания на валидационной выборке для лучшей модели
predicted_valid_DTC = best_model_DTC.predict(features_valid)
probabilities_valid_DTC = best_model_DTC.predict_proba(features_valid)
probabilities_one_valid_DTC = probabilities_valid_DTC[:, 1]

# подсчет показателей для лучшей модели
f1_valid_DTC = f1_score(target_valid, predicted_valid_DTC)
re_score_DTC = recall_score(target_valid, predicted_valid_DTC)
pre_DTC = precision_score(target_valid, predicted_valid_DTC)
auc_roc_DTC = roc_auc_score(target_valid, probabilities_one_valid_DTC)

# Вывод результатов
print("-"*100)
print("Best f1:", best_f1_DTC)
print("Best max depth:", best_depth_DTC)
print("Best model DTC:", best_model_DTC)

# вывод результатов показателей для DTC
print("-"*100)
print('DecisionTreeClassifier:')
print('F1:', np.round(f1_valid_DTC, 3))
print('Recall:', np.round(re_score_DTC, 3))
print('Precision:', np.round(pre_DTC, 3))
print('AUC ROC:', np.round(auc_roc_DTC, 3))
print("-"*100)
```

### RandomForestClassifier

Создадим модель Случайного леса с перебором параметров **n_estimators** и **max_depth**

```python
# диапазон значений depth_values_RFC для перебора в параметре max_depth
depth_values_RFC = range(1, 21)

# диапазон значений est_values_RFC для перебора в параметре n_estimators
est_values_RFC = range(10, 51, 10)

# зададим начальные значения для переменных
best_f1_RFC = 0
best_depth_RFC = 0
best_est_RFC = 0

for depth in depth_values_RFC:
    for esti in est_values_RFC:
        model_RFC = RandomForestClassifier(max_depth=depth, n_estimators=esti, random_state=112)
        model_RFC.fit(features_train, target_train)
        target_pred = model_RFC.predict(features_valid)
        f1_RFC = f1_score(target_valid, target_pred)
        # print('max_depth:', depth, 'n_estimators:', esti, 'f1:', f1_RFC)
        if f1_RFC > best_f1_RFC:
            best_f1_RFC = f1_RFC
            best_depth_RFC = depth
            best_est_RFC = esti
            
# обучение модели случайного леса с лучшими параметрами max_depth и n_estimators
best_model_RFC = RandomForestClassifier(max_depth=best_depth_RFC, n_estimators=best_est_RFC, random_state=112)
best_model_RFC.fit(features_train, target_train)
            
# предсказания на валидационной выборке для лучшей модели
predicted_valid_RFC = best_model_RFC.predict(features_valid)
probabilities_valid_RFC = best_model_RFC.predict_proba(features_valid)
probabilities_one_valid_RFC = probabilities_valid_RFC[:, 1]

# подсчет показателей для лучшей модели
f1_valid_RFC = f1_score(target_valid, predicted_valid_RFC)
re_score_RFC = recall_score(target_valid, predicted_valid_RFC)
pre_RFC = precision_score(target_valid, predicted_valid_RFC)
auc_roc_RFC = roc_auc_score(target_valid, probabilities_one_valid_RFC)            
            
# вывод результатов показателей для RFC
print("-"*100)
print('Best F1_RFC:', np.round(best_f1_RFC, 3))
print('Best depth RFC:', best_depth_RFC)
print('Best estimators RFC:', best_est_RFC)
print('Best model RFC:', best_model_RFC)
print("-"*100)

# вывод результатов показателей для RFC
print("-"*100)
print('RandomForestClassifier:')
print('F1:', np.round(f1_valid_RFC, 3))
print('Recall:', np.round(re_score_RFC, 3))
print('Precision:', np.round(pre_RFC, 3))
print('AUC ROC:', np.round(auc_roc_RFC, 3))
print("-"*100)
```

```python
# На вход ROC-кривая принимает значения целевого признака и вероятности положительного класса
# Перебирает разные пороги и возвращает три списка: значения FPR, значения TPR и рассмотренные пороги
fpr_RFC, tpr_RFC, thresholds_RFC = roc_curve(target_valid, probabilities_one_valid_RFC)

# построим график
plt.figure(figsize=(5, 5))

# ROC-кривая случайной модели
plt.plot([0, 1], [0, 1], linestyle='--')
plt.plot(fpr_RFC, tpr_RFC)

# установка границ осей от 0 до 1
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])

# подпись осей
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')

# заголовок к графику "ROC-кривая"
plt.title('ROC-кривая для RandomForestClassifier')

plt.show()
```

### Увеличение выборки Upsampling

Чтобы сделать объекты редкого класса не такими редкими в данных самые важные задачи повторяются по несколько раз, чтобы их легче запомнить. Такая техника называется **upsampling** (от англ. up, «вверх»; sampling, «выборка»).

*Преобразование проходит в несколько этапов:*
- Разделение обучающей выборки на отрицательные и положительные объекты;
- Копирование несколько раз положительные объекты;
- С учётом полученных данных создается новая обучающая выборка;
- Перемешивание данных: идущие друг за другом одинаковые вопросы не помогут обучению.
- Копирование объектов несколько раз поможет синтаксис умножения списков в Python. Чтобы повторить элементы списка, он умножается на число (нужное количество раз)

```python
## функция Upsampling
def upsampling (features_train, target_train, repeat):
    
    # разделяем выборку по значениям целевой функции
    features_one = features_train[target==1]
    features_null = features_train[target==0]
    target_one = target_train[target==1]
    target_null = target_train[target==0]

    # увеличиваем и соединяем обратно
    upsampling_features = pd.concat([features_null] + [features_one] * repeat)
    upsampling_target = pd.concat([target_null] + [target_one] * repeat)
    
    # перемешиваем
    upsampling_features, upsampling_target = shuffle(upsampling_features, upsampling_target, random_state=112)
    
    return upsampling_features, upsampling_target

features_balance_up, target_balance_up = upsampling(features_train, target_train, 4)

# проверим баланс
print("Выборка features_balance_up:", features_balance_up.shape)
print("Выборка target_balance_up:", target_balance_up.shape)
```

### Уменьшение выборки Downsampling

Чтобы сделать объекты частого класса не такими частыми, нужно вместо повторения важных вопросов убрать часть неважных. Это можно сделать техникой **downsampling** (от англ. down — «вниз»; sampling — «выборка»).
Преобразование проходит в несколько этапов:
- Разделение обучающей выборки на отрицательные и положительные объекты;
- Случайным образом отбрасывается часть из отрицательных объектов;
- С учётом полученных данных создается новая обучающая выборка;
- Перемешивание данных. Положительные не должны идти следом за отрицательными: алгоритмам будет сложнее обучаться.

Чтобы выбросить из таблицы случайные элементы, применяется функция **sample()**. На вход она принимает аргумент **frac** (от англ. fraction, «доля»). Возвращает случайные элементы в таком количестве, чтобы их доля от исходной таблицы была равна frac.

```python
def downsampling(features_train, target_train, fraction):
    features_null = features_train[target_train==0]
    features_one = features_train[target_train==1]
    target_null = target_train[target_train==0]
    target_one = target_train[target_train==1]
    
    downsampling_features = pd.concat([features_null.sample(frac=fraction, random_state=112)] + [features_one])
    downsampling_target = pd.concat([target_null.sample(frac=fraction, random_state=112)] + [target_one])
    
    downsampling_features, downsampling_target = shuffle(downsampling_features, downsampling_target, random_state=112)
    return downsampling_features, downsampling_target

features_balance_dw, target_balance_dw = downsampling(features_train, target_train, 0.25)

# проверим баланс
print("Выборка features_balance_dw:", features_balance_dw.shape)
print("Выборка target_balance_dw:", target_balance_dw.shape)
```
