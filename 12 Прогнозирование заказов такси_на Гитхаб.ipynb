````python
def df_info(df):
    '''
    функиия для вывода основных показателей датафрейма
    
    '''
    print("-"*100)
    print('Общая информамия о датафрейме:')
    print("-"*100)
    print(df.info())
    
    print("-"*100)
    print('Первые 5 строк таблицы')
    print("-"*100)
    display(df.head(5))
    
    print("-"*100)
    print('Количество пропусков')
    print("-"*100)
    print(df.isnull().sum().sort_values(ascending=False))
````

````python
def model_fit(model, X_train, y_train):
    start = time.time()
    '''
    функция для обучения модели и расчета времени
    
    '''
    model.fit(X_train, y_train)
    end = time.time()
    time_fit = end - start
    return time_fit, model
````

````python
def model_predict(model, X_test):
    start = time.time()
    '''
    функция для предсказания модели и расчета времени
    
    '''
    predict = model.predict(X_test)
    end = time.time()
    time_predict = end - start
    return time_predict, predict
````

````python
def model_metrics(RMSE, time_fit, pred_time, model_name, data):
    '''
    функция добавит в результат по модели: RMSE, время обучения и время предсказания
    
    '''
    data.loc[model_name,'RMSE'] = np.round(RMSE, 2)
    data.loc[model_name,'fit_time'] = np.round(time_fit, 2)
    data.loc[model_name,'predict_time'] = np.round(pred_time, 2)
    data.loc[model_name,'sum_time'] = np.round((pred_time + time_fit), 2)
    data.loc[model_name,'Model'] = model_name
    return data
````

````python
def RMSE(target, predict):
    '''
    функиия для расчета метрики RMSE
    
    '''
    return (mean_squared_error(target, predict))**0.5
````

### Создание признаков

````python
# сбросим сезонность 
df = df.resample('1H').sum()
````

Создадим признаки для горизонта прогнозирования в один шаг.
Напишем функцию для создания признаков.
1. **Календарные признаки** (calendar features)
Во многих данных тренды и сезонность привязаны к конкретной дате. Тип datetime64 в Pandas уже содержит нужную информацию, осталось лишь представить её как отдельные столбцы.
2. **«Отстающие значения»** (lag features)
Предыдущие значения временного ряда подскажут, будет ли функция x(t) расти или уменьшаться. Получим отстающие значения знакомой функцией shift():
Для первых дат есть не все отстающие значения, поэтому в этих строках стоят NaN.
3. **Скользящее среднее**
Скользящее среднее как признак задаёт общий тренд временного ряда. Повторим, как его вычислять:
Скользящее среднее в моменте t учитывает текущее значение ряда x(t). Это некорректно: целевой признак «убежал» в признаки. Вычисление скользящего среднего не должно включать в себя текущее значение ряда.

````python
# Создадим дополнительные признаки для обучения модели
def make_features(df, column, MAX_LAG, ROLLING_MEAN_SIZE):
    '''
    функиия для создания дополнительных признаков
    
    '''
    df['dayofweek'] = df.index.dayofweek
    df['day'] = df.index.day
    df['hour'] = df.index.hour
    
    for lag in range(1, MAX_LAG + 1):
        df['lag_{}'.format(lag)] = df[column].shift(lag)

    df['rolling_mean'] = df['num_orders'].shift().rolling(ROLLING_MEAN_SIZE).mean()
    return df
````

````python
# вызовим функцию make_features для создания новых признаков с аргументами:
# датафрейм, признак с заказами, максимальный лаг, ширина окна лага
make_features(df, 'num_orders', MAX_LAG, ROLLING_MEAN_SIZE)
df.head(5)
````

### LinearRegression

Обучим модель линейной регрессии на закодированных значениях признаков и оценим ее производительность на тренировочных данных.

Сначала создадим объект класса `LinearRegression()` и передадим в функцию `model_fit` вместе с обучающими данными **X_train** и **y_train**. Функция `model_fit` возвращает время обучения модели и саму модель.

Затем модель используется для предсказания значений на тестовых данных с помощью функции `model_predict`, которая возвращает время предсказания и предсказанные значения.

Далее вычисляется среднеквадратичное отклонение между предсказанными и реальными значениями с помощью функции `mean_squared_error` из библиотеки **numpy**. Результат сохраняется в переменную **LR_RMSE**.

Наконец, результаты оценки модели (**RMSE**, время обучения и время предсказания) передаются в функцию `model_metrics`, которая добавляет их в словарь `model_result` под названием **Linear regression**.

````python
RMSE_score = make_scorer(RMSE, greater_is_better=False)
````

````python
%%time
#модель линейной регрессии
cv = TimeSeriesSplit(n_splits=3).split(X_train)
model_lr = LinearRegression() 
model_lr.fit(X_train, y_train)
score_lr = cross_val_score(model_lr, X_train, y_train, scoring='neg_mean_squared_error', cv=cv)

print('Качество модели LinearRegression на кросс-валидации:', round(np.mean(-score_lr)**0.5, 2))
````
